Dans cette étude nous allons nous intéresser aux données recueillis sur les différentes mesures, de la qualité de l'air et sur les concentrations de plusieurs particules, effectuées en Bretagne sur l'ensemble de l'année 2020.   

L'ensemble des données sont accessibles sur le site de [data.airbreizh](https://data.airbreizh.asso.fr/contenu/services_didon.html#).   

Cette étude commandée par la région Bretagne vise à étudier l'impact du confinement lié à la pandémie de COVID-19 sur la pollution atmosphérique dans la région.  

- **Les différents indicateurs:**

    - `valeur` : Valeur de l'indice, compris entre 1 et 10 ou nul si absent, de la qualité de l'air
    - `val_no2` : Valeur du sous-indice de dioxyde d'azote (NO2), entier de 1 à 10 ou 0 si absent ou nul si non calculé
    - `val_so2` : Valeur du sous-indice de dioxyde de souffre (SO2), entier de 1 à 10 ou 0 si absent ou nul si non calculé
    - `val_o3` : Valeur du sous-indice d'ozone (O3), entier de 1 à 10 ou 0 si absent ou nul si non calculé
    - `val_pm10` : Valeur du sous-indice de particules fines de diamètre inférieur à 10µm (PM10), entier de 1 à 10 ou 0 si absent ou nul si non calculé
    - `val_pm25` : Valeur du sous-indice de particules fines de diamètre inférieur à 25µm (PM25), entier de 1 à 10 ou 0 si absent ou nul si non calculé

- **Les différentes zones de la région Bretagne:**

    - `Rennes Métropole`
    - `Brest Métropole`
    - `CA Quimper Bretagne Occidentale`
    - `CA Lorient`
    - `CA Golfe du Morbihan - Vannes`
    - `CA Saint-Brieuc Armor`
    - `CA du Pays de Saint-Malo`

En observant la première figure au regard de l'évolution des sous-indices `O3` et `PM10`, on observe deux phénomènes intéressants.

- **PM10 :** Une baisse significative des particules fines PM10 à partir du mois de Mai. Du fait de la baisse d'activité dès Mars le nombre de fines particules stockées dans l'air à progressivement diminué. Cela n'est pas instantané, les particules mettent un certain temps à disparaître.

- **O3 :** Une hausse du sous-indice O3 dès le mois d'avril indiquerait que la disparition des fines particules PM10 permettrait aux rayons du soleil de mieux pénétrer l'atmosphère et donc engendrerait uen hause de la concentration due à des temps ensoleillé plus long.

**Ainsi, le confinement aurait un effet bénéfique global sur la qualité de l'air, bien qu'il ne faut pas négliger l'importance de la concentration de l'O3 qui peut avoir des effets néfastes sur la santé à concentrations trop élevées. **
